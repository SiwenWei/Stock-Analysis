```python
days = 365
dt = 1/days
mu = returns.mean()['SU']
sigma = returns.std()['SU']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    price=np.zeros(days)
    price[0]=start_price
    
    shock=np.zeros(days)
    drift=np.zeros(days)
    
    #to run the price for a number of days
    for x in range(1,days):
        shock[x]=sigma*np.sqrt(dt)*np.random.normal(mu,sigma) #normal distribution different from the lecture
        drift[x]=mu*dt
        price[x]=price[x-1]+(price[x-1]*(drift[x]+shock[x]))
    return price
        
```

```python
for run in range(300):
    plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
    
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Analysis for Suncor')
```

```python
runs=10000

simulation=np.zeros(runs)

for run in range(runs):
    simulation[run]=stock_monte_carlo(start_price,days,mu,sigma)[days-1]
```

```python
q=np.percentile(simulation,1)
q
```

```python
plt.hist(simulation,bins=400)

#to draw a histogram to show all the final prices of Google

# Using plt.figtext to fill in some additional information onto the plot

# Starting Price
plt.figtext(0.6, 0.8, s="Start price: $%.2f" %start_price)
# Mean ending price
plt.figtext(0.6, 0.7, "Mean final price: $%.2f" % simulation.mean())

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6, 0.6, "VaR(0.99): $%.2f" % (start_price - q,))

# Display 1% quantile
plt.figtext(0.15, 0.6, "q(0.99): $%.2f" % q)

# Plot a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# Title
plt.title(u"Final price distribution for Suncor Stock after %s days" % days, weight='bold');
```
