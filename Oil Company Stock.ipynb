```python
closing_df=df['Adj Close']
closing_returns=closing_df.pct_change()
```

```python
closing_returns.head()
```

```python
sns.jointplot('SU','IMO',closing_returns)
```

```python
sns.pairplot(closing_returns.dropna())
```

```python
returns_fig=sns.PairGrid(closing_returns.dropna())
returns_fig.map_upper(plt.scatter,color='indianred')
returns_fig.map_diag(plt.hist,bins=50)
returns_fig.map_lower(sns.kdeplot,cmap='cool_d')
```

```python
returns_fig=sns.PairGrid(closing_df)
returns_fig.map_upper(plt.scatter,color='indianred')
returns_fig.map_diag(plt.hist,bins=50)
returns_fig.map_lower(sns.kdeplot,cmap='cool_d')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(closing_returns.dropna().corr(),annot=True,center=0,linewidth=0.5,cmap='YlGnBu')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(closing_df.corr(),annot=True,center=0,linewidth=0.5,cmap='YlGnBu')
```

```python
returns = closing_returns.dropna()
returns.head()
```

```python
returns.mean()
```

```python
returns.std()
```

```python
area=np.pi*20
plt.scatter(returns.mean(),returns.std(),s=area)
plt.xlabel('Expected Return')
plt.ylabel('Risk')

#label the scatter plot by using annotation to annotate axes and points
for label, x, y in zip(returns.columns,returns.mean(),returns.std()):
    plt.annotate(label,xy=(x,y),xytext=(50,50),
                textcoords='offset points',
                arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-0.3'))
```
